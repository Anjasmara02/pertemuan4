```python
import seaborn as sns
```

```python
df_penguin = sns.load_dataset('penguins')
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
print("jumlah secord yang punya nilai NULL:")
record_null = df_penguin.isna().sum()
print(record_null)
```

```python
df_penguin['bill_length_mm'].fillna(
    df_penguin["bill_depth_mm"].mean(),inplace=True
)
df_penguin
```

simplelputer

```python
imputer = SimpleImputer(strategy='mean')
df_penguin["bill_depth_mm"] = imputer.fit_transform(
    df_penguin[["bill_depth_mm"]]
)
df_penguin
```

```python
df_penguin.penguin.iloc[1,:] =df_penguin.iloc[0,:]
```

duplicated

```python
df_penguin.duplicated().sum()
```

```python
df_penguin.drop_duplicates(inplace=True)
```

```python
df_penguin.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# normalisasi = minmaxscaler (bill_length_mm)
nrm = MinMaxScaler()
df_penguin["bill_length_mm"] = nrm.fit_transform(
    df_penguin[["bill_length_mm"]]
)
```

```python
std =StandardScaler()
df_penguin["bill_depth_mm"] = std.fit_transform(
    df_penguin[["bill_depth_mm"]]
)
```

```python
df_penguin['bill_depth_mm'].min()
```

```python
print(df_penguin['bill_depth_mm'].mean())
print(df_penguin['bill_depth_mm'].std())
```

```python
#ordinal encoder
df_penguin["sex"] = df_penguin["sex"].fillna("male")
df_penguin
```

```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
```

```python
#ordinal encoder
ordinal = OrdinalEncoder()
```

```python
df_penguin['sex'] =  ordinal.fit_transform(df_penguin[['sex']])  
```

```python
onehot = OneHotEncoder(sparse=False)
species_enc = onehot.fit_transform(df_penguin[["species"]])
species_enc = pd.DataFrame(species_enc)
```

```python
df_penguin["Adelie"] = species_enc[0]
df_penguin["Gento"] = species_enc[1]
df_penguin["Chistrap"] = species_enc[2]
```
